```
def dt_to_number(x):
    y = int(x[2:4])
    m = int(x[5:7])
    return (y - 14)* 12 + m
```

```
import matplotlib.pyplot as plt


for i in (df_e_y_all_0_index):

    fr_number = dt_to_number(df_train[df_train["fr_yn"] == "Y"]["dt_of_fr"].loc[i])
    plt.figure()
    plt.plot([x for x in range(60)],df_e_y.loc[i])
    plt.axvline(x=fr_number, color='r', linestyle='--', linewidth=3)
    plt.show()
```
